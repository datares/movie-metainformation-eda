```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
movies_data = pd.read_csv('Datasets/movies_metadata.csv')
```

# Cleaning the Data

First we looked at the metainformation on the movie data metainformation.

```python
movies_data.info()
```

```python
movies_data.drop(columns=['homepage', 'tagline', 'video', 'overview', 'poster_path', 'imdb_id'], inplace=True)
```

```python
movies_data.isnull().sum()
```

We dropped rows that had no title, and changed budget to be an int.

```python
movies_data.dropna(subset=['title'], inplace=True)
```

```python
movies_data = movies_data.astype({'budget': 'int64'})
```

```python
movies_data[movies_data['budget'] == 0]
```

Next we checked the numerical columns to ensure there were no out of place minimum and maximum values.

```python
movies_data.describe()
```

```python
movies_data.loc[movies_data['budget'] == 0,'budget'] = np.nan
movies_data.loc[movies_data['runtime'] == 0,'runtime'] = np.nan
movies_data.loc[movies_data['revenue'] == 0,'revenue'] = np.nan
```

```python
movies_data[movies_data['runtime'] < 40].count()
```

According to [the academy of Feature Films, a movie is at least 40 minutes long](https://en.wikipedia.org/wiki/Feature_film#:~:text=According%20to%20the%20Academy%20of,75%20and%20210%20minutes%20long.) so we dropped all movies with a runtime less than 40 minutes (since those are more like shorts).

```python
movies_data.drop(movies_data[movies_data['runtime'] < 40].index, inplace = True)
```

```python
movies_data = movies_data[['id', 'title', 'genres']]
movies_data
```

Finally we exported our cleaned data as a csv.

```python
movies_data.to_csv('cleaned-movies-metainformation.csv', index=False)  
```
