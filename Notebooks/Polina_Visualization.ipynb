```python
import pandas as pd
from pandas import DataFrame
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
movie_data = pd.read_csv('movie_data.csv')
```

```python
revenue = DataFrame(movie_data, columns = ['revenue'])
runtime = DataFrame(movie_data, columns = ['runtime'])
```

```python
plt.figure(figsize=(20,12))
plt.scatter(runtime, revenue, alpha = 0.3, color = 'purple')
plt.title('Film Runtime vs Film Revenue')
plt.xlabel('Runtime (minutes)')
plt.ylabel('Revenue ($)')
plt.xlim(0, 400)
plt.ylim(0, 2900000000)
plt.show()
```
