```python
movies_data = movies_data[['id', 'runtime', 'budget']]
```

```python
movies_data = movies_data.merge(credits_data,on='id')
movies_data
```

```python
movies_data['crew']
```

```python
def makeColumn(series, key):
    count = 0
    col = []
    string = ""
    for lis in series:
        li = lis.strip('][').split(', ')
        for dic in li:
            di = dic.strip('}{').replace("'", '').split(':')
            if di[0] == key:
                string += di[1]
                string += ","
        col.append(string)
        string = ""
    return pd.Series(col)
```

```python
movies_data['cast'] = makeColumn(movies_data['cast'], 'name')
```

```python
movies_data['crew_gender'] = makeColumn(movies_data['crew'], 'gender')
```
